```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_linnerud
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the Linnerud dataset
linnerud = load_linnerud()

# Extract waistline data and number of situps accomplished
X_waistline = linnerud.data[:, 0].reshape(-1, 1)  # Reshape to ensure it's a 2D array
y_situps = linnerud.target[:, 0]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_waistline, y_situps, test_size=0.2, random_state=42)

# Choose a regression model (Linear Regression)
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r_squared = model.score(X_test, y_test)

print("Mean Squared Error:", mse)
print("R-squared:", r_squared)

# Plot the relationship
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=4, label='Predicted')
plt.xlabel('Waistline')
plt.ylabel('Number of Situps')
plt.title('Relationship between Waistline and Number of Situps')
plt.legend()
plt.show()
```
